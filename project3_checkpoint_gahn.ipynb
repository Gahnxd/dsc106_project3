```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
male_act = pd.read_csv('Male_Act.csv')
male_temp = pd.read_csv('Male_Temp.csv')
fem_act = pd.read_csv('Fem_Act.csv')
fem_temp = pd.read_csv('Fem_Temp.csv')
```

```python
fem_act['avg'] = fem_act.mean(axis=1)
fem_act['hour'] = fem_act.index.values // 60
fem_act['day'] = fem_act['hour'] // 24
fem_act['light'] = ((fem_act['hour'] // 12) % 2) == 1
fem_act['estrus_start'] = fem_act['day'].isin([i for i in range(1, 14, 4)])
fem_act = np.cumsum(fem_act.groupby('day')['estrus_start'].mean()).reset_index().rename(columns={'estrus_start': 'estrus_cycle'}).merge(fem_act, on='day')

fem_temp['avg'] = fem_temp.mean(axis=1)
fem_temp['hour'] = fem_temp.index.values // 60
fem_temp['day'] = fem_temp['hour'] // 24
fem_temp['light'] = ((fem_temp['hour'] // 12) % 2) == 1
fem_temp['estrus_start'] = fem_temp['day'].isin([i for i in range(1, 14, 4)])
fem_temp = np.cumsum(fem_temp.groupby('day')['estrus_start'].mean()).reset_index().rename(columns={'estrus_start': 'estrus_cycle'}).merge(fem_temp, on='day')

male_act['avg'] = male_act.mean(axis=1)
male_act['hour'] = male_act.index.values // 60
male_act['day'] = male_act['hour'] // 24
male_act['light'] = ((male_act['hour'] // 12) % 2) == 1
male_act = np.cumsum(fem_act.groupby('day')['estrus_start'].mean()).reset_index().rename(columns={'estrus_start': 'estrus_cycle'}).merge(male_act, on='day')

male_temp['avg'] = male_temp.mean(axis=1)
male_temp['hour'] = male_temp.index.values // 60
male_temp['day'] = male_temp['hour'] // 24
male_temp['light'] = ((male_temp['hour'] // 12) % 2) == 1
male_temp = np.cumsum(fem_temp.groupby('day')['estrus_start'].mean()).reset_index().rename(columns={'estrus_start': 'estrus_cycle'}).merge(male_temp, on='day')
```

```python
fig = px.line(fem_temp.groupby('hour')[['avg']].mean(), y='avg')
fig.add_scatter(x=male_temp.groupby('hour')[['avg']].mean().index, y=male_temp.groupby('hour')[['avg']].mean()['avg'], mode='lines', name='Male')
fig.update_xaxes(
    title='Hour',
)
fig.update_yaxes(
    title='Temperature (°C)',
)
fig.update_layout(
    title=dict(
        text='Average Temperature by Hour',
        x=0.5,
    ),
    width=1400,
    height=600,
)
fig.data[0].name = 'Female'
fig.data[0].showlegend = True

# add vertical lines for each estrus cycle
for i in range(1, 14, 4):
    fig.add_vline(x=i*24, line_width=3, line_dash='dash', line_color='purple', annotation_text='Estrus Cycle Start', annotation_position='top right', annotation_font_size=12)
    
# switch legend color
fig.update_legends(traceorder='reversed')

fig.show()
```

```python
fig = px.line(fem_act.groupby('hour')[['avg']].mean(), y='avg', labels={'avg': 'Female'})
fig.add_scatter(x=male_act.groupby('hour')[['avg']].mean().index, y=male_act.groupby('hour')[['avg']].mean()['avg'], mode='lines', name='Male')
fig.update_xaxes(
    title='Hour',
)
fig.update_yaxes(
    title='Activity',
)
fig.update_layout(
    title=dict(
        text='Average Activity by Hour',
        x=0.5,
    ),
    width=1400,
    height=600,
)
fig.data[0].name = 'Female'
fig.data[0].showlegend = True

# add vertical lines for each estrus cycle
for i in range(1, 14, 4):
    fig.add_vline(x=i*24, line_width=3, line_dash='dash', line_color='purple', annotation_text='Estrus Cycle Start', annotation_position='top right', annotation_font_size=12)

fig.show()
```

```python
import plotly.graph_objects as go
radial_fem = fem_temp.groupby('hour')[['avg']].mean().reset_index()
radial_fem['theta'] = radial_fem['hour'] / len(radial_fem['hour']) * 360
radial_male = male_temp.groupby('hour')[['avg']].mean().reset_index()
radial_male['theta'] = radial_male['hour'] / len(radial_male['hour']) * 360

fig = px.line_polar(radial_fem, r='avg', theta='theta')
fig.add_trace(go.Scatterpolar(r=radial_male['avg'], theta=radial_male['theta'], mode='lines', name='Male', cliponaxis=False))
fig.update_layout(
    title=dict(
        text='Average Temperature by Hour',
        x=0.5,
    ),
    width=800,
    height=600,
)
fig.update_polars(radialaxis=dict(
    visible=True,
    range=[35, 39],
))

for i in range(1, 14, 4):
    fig.add_trace(go.Scatterpolar(
        r=[35, 39],
        theta=[(i*24) / len(radial_fem['hour']) * 360, (i*24) / len(radial_fem['hour']) * 360],
        mode='lines',
        line=dict(
            color='purple',
            width=3,
            dash='dash',
        ),
        name='Estrus Cycle Start',
        showlegend=False,
    ))
fig.data[0].name = 'Female'
fig.data[0].showlegend = True
fig.data[-1].showlegend = True

fig.update_layout(
    polar=dict(
        angularaxis=dict(
            tickmode='array',
            tickvals=[round(i/len(radial_fem['hour'])*360) for i in range(0, max(radial_fem['hour']), 12)],
            ticktext=[f'{i}' for i in range(0, max(radial_fem['hour']), 12)],
        ),
    ),
)

# add labels for axis
fig.add_annotation(
    x=1,
    y=0.15,
    xref='paper',
    yref='paper',
    text='Hour',
    showarrow=False,
    font=dict(
        size=14,
    ),
)

fig.show()
```
